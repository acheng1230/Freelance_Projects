```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url = 'https://connect.iowaleague.org/iloc/Contact_Management/CityDirectory.aspx'
```

```python
driver = webdriver.Firefox()
```

```python
driver.get(url)
```

```python
df = pd.read_csv("iowa-table.csv")
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
address_id = "ctl01_TemplateBody_WebPartManager1_gwpciMiniProfile_ciMiniProfile_contactAddress__address"
phone_id = "ctl01_TemplateBody_WebPartManager1_gwpciMiniProfile_ciMiniProfile_contactAddress__phoneNumber"
```

```python
address_list = []
#phone_list = []

for link in df['URL']:
    driver.get('https://connect.iowaleague.org/' + link)
    
    try:
        address = driver.find_element(by='id', value=address_id).text
        #phone = driver.find_element(by='id', value=phone_id).text
    except NoSuchElementException:
        pass
    
    address_list.append(address)
    #phone_list.append(phone)
    
    # driver.implicitly_wait(2) # seconds
```

```python
df["Address"] = address_list
```

```python
driver.quit()
```
